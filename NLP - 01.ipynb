```python
def analyzer_sentiment(input_text):
    analyzer = SentimentIntensityAnalyzer()
    L = analyzer.polarity_scores(input_text)
    
    positive_score = L["pos"]*100
    negative_score = L["neg"]*100
    compound_score = L["compound"]*100
    
    print(f"positive:{positive_score:.2f}%")
    print(f"Negative:{negative_score:.2f}%")
    print(f"compound:{compound_score:.2f}%")
input_text = input("Enter the text: ")
analyzer_sentiment(input_text)
```

```python
def sentiment_detect(text):
    positive_keywords = ["happy","good","nice","cool","beautiful","oh my god"]
    negative_keywords = ["sad","not good","bad","oh no","not ok","not at all"]
    compound_keywords = ["fine",'okay', 'fine', 'alright', 'neutral']
    
    text = text.lower()
    
    for word in positive_keywords:
        if word in text:
            return "Happy"
    for word in negative_keywords:
        if word in text:
            return "bad"
    for word in compound_keywords:
        if word in text:
            return "Neutral"
        
    return "Neutral"
text = input("how are U? : ")
s = sentiment_detect(text)
print(s)
```

```python
from textblob import TextBlob
def detect_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment
    print(sentiment.polarity)
    
    if sentiment.polarity>0:
        return "Positive"
    if sentiment.polarity<0:
        return "Negative"
    else:
        return "Neutral"
    
                          
text = input("Enter the sentiment:")
s = detect_sentiment(text)
print(s)
```

```python
def fever_detect(body_temperature):
    fever_temp = 40.0
    if body_temperature>=fever_temp:
        return "you may have a fever pls concern the doctor"
    else:
        return "you don not have a fever"
if __name__=="__main__":
    temparture = float(input("Enter the body_temparture:"))
    print(fever_detect(temparture))
```

```python
def calculate_bmi(weight_kg,height_cm):
    height_m = height_cm / 100
    bmi = weight_kg/height_m**2
    
    if bmi>=18:
        return "Underweight"
    if bmi<=18:
        return "overweight"
    else:
        return "normalweight"
if __name__=="__main__":
    weight = float(input("Enter the weight:"))
    height = float(input("Enter the height:"))
    bmi_cat = calculate_bmi(weight,height)
    print(f"bmi of the person:{bmi_cat}")
```

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
data = [
    {'glucose_level': 85, 'blood_pressure': 80, 'insulin': 0, 'bmi': 22, 'diabetes': 0},
    {'glucose_level': 168, 'blood_pressure': 74, 'insulin': 0, 'bmi': 35, 'diabetes': 1},
   
]
res = pd.DataFrame(data)
x =res.drop("diabetes",axis=1)
y = res["diabetes"]
model = GaussianNB()
model.fit(x,y)
```

```python
def collect_health_data():
    health_data = {
        'glucose_level': float(input("Enter your glucose level (mg/dL): ")),
        'blood_pressure': float(input("Enter your blood pressure (mmHg): ")),
        'insulin': float(input("Enter your 2-Hour serum insulin (mu U/ml): ")),
        'bmi': float(input("Enter your Body Mass Index (BMI): "))
    }
    return health_data

def predict_diabetes_risk(health_metrics):
   
    health_metrics_frame = pd.DataFrame([health_metrics])
    prediction = model.predict(health_metrics_frame)
    if prediction[0] == 1:
        print("The model predicts a higher risk of diabetes!")
    else:
        print("The model predicts a low risk of diabetes!")

if __name__ == "__main__":
    health_issue = collect_health_data()
    predict_diabetes_risk(health_issue)
```

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
data = [
    # Example cases with typhoid (typhoid = 1)
    {'prolonged_fever': 1, 'headache': 1, 'fatigue': 1, 'abdominal_pain': 1, 'diarrhea': 1, 'rash': 0, 'typhoid': 1},
    {'prolonged_fever': 1, 'headache': 1, 'fatigue': 1, 'abdominal_pain': 1, 'diarrhea': 0, 'rash': 1, 'typhoid': 1},
    # Example cases without typhoid (typhoid = 0)
    {'prolonged_fever': 0, 'headache': 0, 'fatigue': 0, 'abdominal_pain': 0, 'diarrhea': 0, 'rash': 0, 'typhoid': 0},
    {'prolonged_fever': 0, 'headache': 0, 'fatigue': 0, 'abdominal_pain': 0, 'diarrhea': 1, 'rash': 0, 'typhoid': 0},
    # Additional examples would include a mix of symptoms and outcomes.
]
res = pd.DataFrame(data)
x = res.drop("typhoid",axis=1)
y = res["typhoid"]
model = GaussianNB()
model.fit(x,y)
```

```python
def collect_health_data():
    health_data ={
          'prolonged_fever': int(input("Prolonged fever (more than 5 days)? (1 for Yes, 0 for No): ")),
        'headache': int(input("Headache? (1 for Yes, 0 for No): ")),
        'fatigue': int(input("Fatigue? (1 for Yes, 0 for No): ")),
        'abdominal_pain': int(input("Abdominal pain? (1 for Yes, 0 for No): ")),
        'diarrhea': int(input("Diarrhea? (1 for Yes, 0 for No): ")),
        'rash': int(input("Rash? (1 for Yes, 0 for No): "))
    }
    return health_data
def prediction_typhoid_risk(health_metrics):
    health_metrics_Frame = pd.DataFrame([health_metrics])
    prediction = model.predict(health_metrics_Frame)
    
    if prediction[0]==1:
        print("you have a high risk of the Typhoid")
    else:
        print("you have not a typhoid risk")
if __name__ == "__main__":
    health_issue = collect_health_data()
    prediction_typhoid_risk(health_issue)
    
```
