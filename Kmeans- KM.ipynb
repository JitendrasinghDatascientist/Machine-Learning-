```python
import matplotlib.pyplot as plt 
import numpy as np
from sklearn.cluster import KMeans
data = ([[1,22,33,44,55],
        [2,23,24,24,25],
        [3,33,44,55,66],
        [4,556,77,899,89],
        [5,67,88,77,66]])
data = np.array(data)
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
labels = kmeans.labels_ 

for i,label in enumerate(labels):
    print({i+1},{label+1})
plt.scatter(data[ :,0],data[ : ,3], c=labels ,cmap="rainbow")
plt.xlabel("hours worked")
plt.ylabel("WFH")
plt.title("employee data")
plt.show()
```
